```python
import numpy as np  
import sys 
from fractions import Fraction as fr
from math import nan
from matplotlib import *
# DISCLAIMER : 
# 1. Input variables should be inserted exactly in the way mentioned in the PDF
# 2. Input given is :
# 3 2
# 1 1 3
# 3 2 1 leq 3
# 2 1 2 leq 2

# To use this file run it and give space seperated inputs for each line. After Input for a line press enter and then input the next line
# Line 1: no of original variables , no. of equations. 
# Line 2: Coefficients of the variables in the objective function (space seperated)
# Line 3 (onwards): describe the constarint equations - first the coefficient of the original variables, then leq/geq and then the value of b:
# All values should be space seperated for line 3 onwards (i.e space seperated for each line. After line press enter to go to next line )

## input manipulation
# n is no of variables and e is no of equations
n,e=input().split(' ')

#2d array to kep track of coefficients
constr=[[0 for i in range(int(n))]for j in range(int(e))]    

#1d array to store b values
type_1=[0 for i in range(int(e))]   

#arg is a temporary list
arg=[]  
arg=input().split(' ')
arg = [int(i) for i in arg]
 
for i in range(int(e)):
    arg.append(0)

obj_coeff=np.array(arg)
arg.clear()
check_flag=0
# construct Matrix for constraints 
for i in range(int(e)):
    arg=input().split(' ')
    for j in range(len(arg)-2) :
        constr[i][j]=float(arg[j])
    type_1[i]=float(arg[len(arg)-1])
    if(arg[len(arg)-2]!="leq"):
        check_flag=1
    arg.clear()
if check_flag==1:
    #print("")
    sys.exit("Wrong Input type ")

# Assignment of arrays
# A = final 2d array to kep track of constraint coefficients  
  
a_val=np.array(constr)
b_val=np.array(type_1)
eye = np.identity(int(e), dtype = float)
A= np.hstack((a_val, eye))
b=b_val
c=obj_coeff
obj_coeff1=obj_coeff.T
arg.clear()

# MR is the array of Min Ratio for the question 
# OP is the string array to maintain operations

MR = np.zeros(int(e))
MR = MR.reshape(int(e),1)
cb = np.array(c[int(n)]) 
for i in range(int(e)-1):
    cb=np.vstack((cb,c[int(n)+i+1]))
B_array=np.array(int(n)+1)
for i in range (int(e)-1):
    B_array=np.vstack((B_array,int(n)+i+2))
         
xb = np.transpose([b])                   
simplex_table = np.hstack((B_array, cb))              
simplex_table = np.hstack((simplex_table, xb))          
 
#simplex_table is final 2d array for simplex calculation
simplex_table = np.hstack((simplex_table, A))   
simplex_table = np.hstack((simplex_table, MR))        
simplex_table = np.array(simplex_table, dtype ='float')  
OP = np.array(["                         " for _ in range(int(e))])

# inputs end 
  
# if its a minimization problem, make this variable =  1 
min_check = 0
 
 
if_ans_reached = 0     
iteration_count = 1
unbounded = 0  
while if_ans_reached == 0: 
  
    i = 0
    
    #calculate relative profit
    relative_profit_array = [] 
    while i<len(A[0]): 
        relative_profit_array.append(np.sum(simplex_table[:, 1]*simplex_table[:, 3 + i])-c[i]) 
        i = i + 1
    
    flag = 0
    for profit in relative_profit_array: 
        if profit<0: 
            flag = 1
            break
# if all values >= 0, optimal solution is acheived 
    if flag == 0: 
        if_ans_reached = 1
    i = 0
    #Print Simplex simplex_table
    print("Iteration Count: ", end =' ') 
    print(iteration_count)
    print("\t\t\tCj",end="\t")
    for j in obj_coeff1:
        print(str(j),end="\t")
    print()
    r = -1
    # entering variable 
    k = relative_profit_array.index(min(relative_profit_array)) 
    min1 = 999999
      
    # (only for valid/positive values) Calculate minimum ratio of in simplex simplex_table 
    while i<len(simplex_table): 
        if (float(simplex_table[:, 2][i])>0 and float(simplex_table[:, 3 + k][i])>0):  
            val = float(simplex_table[:, 2][i])/float(simplex_table[:, 3 + k][i]) 
            simplex_table[:,3+len(c)][i]=val
            if val<min1: 
                min1 = val 
                r = i     # leaving variable 
        else:
            simplex_table[:,3+len(c)][i] = -1
        i+= 1
        
    #find pivot_element     
    pivot_element = float(simplex_table[r][3 + k])
    
    str1="Cb"+"\t"+"B"+"\t"+"Xb "+ "\t" + "b"
    for i in range(c.size):
        str1+="\t"+"a"+str(i+1)
    str1+="\t"+"M.R."+"\t"+"Row O/P"
    print(str1)
    
    # Print simplex_table and output formatting
    j_count=0
    for curr_row in simplex_table: 
        if if_ans_reached==1: 
            for el in range(len(curr_row)-1):
                if el == 0:                   
                    if(curr_row[1]<0 and curr_row[1]!=int(round(curr_row[1],2))):
                        print(fr(curr_row[1]).limit_denominator(200), end ='\t')
                    else :
                        print(fr(curr_row[1]).limit_denominator(200), end ='\t')
                    print('a'+str(int(round(curr_row[el],2))), end ='\t')
                    print('x'+str(int(round(curr_row[el],2))), end ='\t')
                    
                    
                elif el>=2:
                    if(curr_row[el]<0 and curr_row[el]!=int(round(curr_row[el],2))):
                        print(fr(curr_row[el]).limit_denominator(200), end ='\t')
                    else :
                        print(fr(curr_row[el]).limit_denominator(200), end ='\t')
            print("\t"+OP[j_count], end="\t") 
            
        if if_ans_reached==0: 
            for el in range(len(curr_row)):
                if el == 0:
                    if(curr_row[1]<0 and curr_row[1]!=int(round(curr_row[1],2))):
                        print(fr(curr_row[1]).limit_denominator(200), end ='\t')
                    else :
                        print(fr(curr_row[1]).limit_denominator(200), end ='\t')
                    print('a'+str(int(round(curr_row[el],2))), end ='\t')
                    print('x'+str(int(round(curr_row[el],2))), end ='\t')
                    
                elif el>=2:
                    if(el==len(curr_row)-1 and curr_row[el]<0):
                        print("N/A",end='\t')
                    elif(curr_row[el]<0 and curr_row[el]!=int(round(curr_row[el],2))):
                        print(fr(curr_row[el]).limit_denominator(200), end ='\t')
                    else:
                        print(fr(curr_row[el]).limit_denominator(200), end ='\t') 
            print(OP[j_count], end="\t")
        j_count+=1
        print() 
    
    #calculate value of Objective Function at this instance
    extra_count=0
    net_sum=0
    while extra_count<len(simplex_table): 
        ind1=int(float(simplex_table[extra_count][0]-1.0))
        ind2=int(float(simplex_table[extra_count][2]))
        net_sum += c[ind1]*ind2 
        extra_count+= 1

    #Print relative profit (zj-cj)
    print("Zj - Cj :  "+"\t\t"+str(net_sum), end ="\t") 
    for curr_profit in relative_profit_array: 
        if(curr_profit!=int(round(curr_profit,2))):
            print(fr(str(curr_profit)).limit_denominator(200), end =",\t")
        else:
            print(fr(str(curr_profit)).limit_denominator(200), end =",\t")
    print() 
    
    i = 0
    trial1=np.array(simplex_table[:, 0])
    trial1= trial1.astype(np.float)
    b_var = trial1
    #check for unbounded problem
    if r ==-1: 
        unbounded = 1
        break
   
    #calculate Row operation array
    while i<int(e): 
        if i != r: 
            OP[i] = "R"+str(i+1)+"' = "+"R"+str(i+1)+" - ("+str(round(simplex_table[i][3 + k],2))+")*R"+str(r+1)+"'"
        else:
            OP[i] ="R"+str(i+1)+"' = "+ "R"+str(r+1)+"/"+str(round(pivot_element,2))
        i += 1
        
    #check for optimum solution
    if flag == 0: 
        print()
        print("The optimum solution is:") 
        break
          
    # pivot row / pivot element (divide) 
    trial1=np.array(simplex_table[r, 2:len(simplex_table[0])])
    trial1= trial1.astype(np.float)    
    simplex_table[r, 2:len(simplex_table[0])] = (trial1 / pivot_element) 
    
    # row modification  
    i = 0
    while i<len(simplex_table): 
        if i != r: 
            simplex_table[i, 2:len(simplex_table[0])] = (simplex_table[i,2:len(simplex_table[0])] - simplex_table[i][3 + k] * simplex_table[r, 2:len(simplex_table[0])]) 
        i += 1
        
    # assign the new basic variable which is entering in place of the leaving variable
    simplex_table[r][0] = (k+1)
    simplex_table[r][1] = (c[k])  
    print()  
    iteration_count+= 1
    
print()  
#For unbounded problem 
if unbounded == 1:  
    sys.exit("Unbounded LPP Problem")
# for cases where solution exists
i = 0
net_sum = 0
array_trial1=[]
array_trial2=[]
while i<len(simplex_table): 
    ind1=int(float(simplex_table[i][0]-1.0))
    ind2=int(float(simplex_table[i][2]))
    array_trial1.append(ind1)
    array_trial2.append(ind2)
    net_sum += c[ind1]*ind2 
    i+= 1
    
#print 0 to n-1 basis variables(original variables) and value of objective function
for ele in range(int(n)): 
    if ele+1 in simplex_table[:,0]:
        flag=array_trial1.index(ele)
        print("x"+str(ele+1)+" = ",str(array_trial2[flag]), end=", ")
    else:
        print("x"+str(ele+1)+" = 0", end=", ")
print("and Z =", end=" ")
    
# if its a minimization problem  
if min_check == 1: 
    print(-fr(str(net_sum)).limit_denominator(200)) 
else: 
    print(fr(str(net_sum)).limit_denominator(200)) 
print()
 
 
```
